```python
#Data Plotting

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("X-Y Graph")
plt.show()


# Bar Chart

import matplotlib.pyplot as plt

subjects = ["Maths", "Physics", "Chem", "CS"]
marks = [80, 75, 85, 90]

plt.bar(subjects, marks)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A','B','C']
values = [3,9,6]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=40)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,40,75,38,66,70]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [25, 45, 55]
labels = ['A', 'B', 'C']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

plt.boxplot(data)
plt.title("Box Plot Example")
plt.xlabel("Dataset")
plt.ylabel("Values")
plt.show()


```
