```python
# LIST OPERATIONS IN ONE PROGRAM

# 1. Creating a list
my_list = [10, 20, 30, 40, 50]
print("Original List:", my_list)

# 2. Accessing elements
print("First element:", my_list[0])
print("Last element:", my_list[-1])

# 3. Slicing
print("Sliced list (1 to 3):", my_list[1:4])

# 4. Adding elements
my_list.append(60)              # Add at end
print("After append:", my_list)

my_list.insert(2, 25)           # Insert at index 2
print("After insert:", my_list)

my_list.extend([70, 80])        # Add multiple elements
print("After extend:", my_list)

# 5. Removing elements
my_list.remove(25)              # Remove specific value
print("After remove:", my_list)

popped = my_list.pop()          # Remove last element
print("Popped element:", popped)
print("After pop:", my_list)

# 6. Updating elements
my_list[0] = 100
print("After update:", my_list)

# 7. Searching
print("Index of 40:", my_list.index(40))
print("Count of 20:", my_list.count(20))

# 8. Sorting
my_list.sort()
print("After sort:", my_list)

# 9. Reversing
my_list.reverse()
print("After reverse:", my_list)

# 10. Length
print("Length of list:", len(my_list))

# 11. Copying
new_list = my_list.copy()
print("Copied list:", new_list)

# 12. Clearing
my_list.clear()
print("After clear:", my_list)
```
