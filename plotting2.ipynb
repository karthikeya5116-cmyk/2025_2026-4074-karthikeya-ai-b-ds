```python
#plotting2
import matplotlib.pyplot as plt
subjects=['maths','physics','chemistry']
marks=[90,85,95]
plt.bar(subjects,marks)
plt.xlabel('subjects')
plt.ylabel('marks')
plt.title('Bar chart')
plt.show()
```
