```python
import pandas as pd

# DataFrame
data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

# Ranking Marks (highest marks = rank 1)
df["Rank"] = df["Marks"].rank(ascending=False)
print(df)
```
