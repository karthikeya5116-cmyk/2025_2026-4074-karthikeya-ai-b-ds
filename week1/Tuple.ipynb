```python
# TUPLE OPERATIONS

# Creating tuples
t = (1,2,3,4,5)
print("Original tuple:", t)

# Accessing elements
print("First element:", t[0])
print("Last element:", t[-1])

# Slicing
print("Sliced tuple:", t[1:4])

# Length
print("Length of tuple:", len(t))

# Membership
print("Is 3 in tuple?", 3 in t)

# Counting elements
print("Count of 2", t.count(2))

# Finding index
print("Index of 4:", t.index(4))

# Looping
print("Tuple elements:")
for item in t:
    print(item)

# Concatenation
t2 = (6,7)
t3 = t + t2
print("After concatenation:", t3)

# Repetition
t4 = t * 2
print("After repetition:", t4)

# Nested tuple
nested = (1, (2, 3), (4, 5))
print("Nested tuple element:", nested[1][0])

# Tuple unpacking
a, b, c, d, e = t
print("Unpacked values:", a, b, c, d, e)

# Converting tuple to list (to modify)
lst = list(t)
lst.append(10)
t_new = tuple(lst)
print("Modified tuple (via list):", t_new)
```
