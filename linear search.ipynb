```python

n = int(input("Enter number of elements: "))
arr = []

# Taking array elements from user
for i in range(n):
    value = int(input(f"Enter element {i+1}: "))
    arr.append(value)

key = int(input("Enter element to search: "))
found = False

# Linear search
for i in range(n):
    if arr[i] == key:
        print("Element found at index", i)
        found = True
        break

if not found:
    print("Element not found")5
```
