```python
import pandas as pd

# Create the data
data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
# Sort rows by Marks (ascending order)
df_sorted_asc = df.sort_values(by="Marks")
print("\nDataFrame sorted by Marks (ascending):")
print(df_sorted_asc)

# Sort rows by Marks (descending order)
df_sorted_desc = df.sort_values(by="Marks", ascending=False)
print("\nDataFrame sorted by Marks (descending):")
print(df_sorted_desc)
```

```python
# Rearrange columns: Marks, Name, Age
df_reordered = df[["Marks", "Name", "Age"]]

print("\nDataFrame with columns rearranged:")
print(df_reordered)
```
