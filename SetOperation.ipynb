```python
# SET OPERATIONS
s1 = {11,22,33,44}
s2 = {33,44,55,66}
print("Union using method:", s1.union(s2))
print("Intersection using method:", s1.intersection(s2))
print("Difference using method:", s1.difference(s2))
print("Symmetric difference using method:", s1.symmetric_difference(s2))
```
