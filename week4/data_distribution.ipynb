```python
#descriptive statistic function
import pandas as pd
data={
    "age":[22,25,47,35,46,50,28,33,39,41],
    "salary":[3000,3200,5500,4000,5200,6000,3500,3700,4500,4800]
}
df=pd.DataFrame(data)
print("Summary Statistics:")
print("mean:",df.mean())
print("median:",df.median())
print("mode:",df.mode())
print("\n minimum value:",df.min())
print("\n maximum value:",df.max())
print("\n standard deviation",df.std())
print("\n variance:",df.var())
print("\n count:",df.count())
print("\n skewness:",df.skew())
print("\n kurtosis:",df.kurt())
```

```python
import pandas as pd
# Creating a sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, None, 35, 40, 28],  # Bob has missing age
    'Salary': [50000, 60000, None, 70000, 62000],  # Charlie has missing salary
    'Allowances': [2000, None, 2000, 2000, 2500]  # Bob has missing allowance
}
df = pd.DataFrame(data)
# 1. Checking for missing values in each column
print("Missing Values Count:\n", df.isnull().sum())
# 2. Dropping rows with any missing values
df_cleaned = df.dropna()
print("\nDataFrame after dropping missing values:\n", df_cleaned)
# 3. Filling missing values with specific values
df_filled = df.fillna({'Age': df['Age'].mean(), 'Salary': df['Salary'].mean(), 'Allowances': 2000})
print("\nDataFrame after filling missing values:\n", df_filled)
```

```python
#data distribution
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

normal_data = np.random.normal(loc=70, scale=5, size=1000)
right_skewed = np.random.exponential(scale=2, size=1000)
left_skewed = -np.random.exponential(scale=2, size=1000)

plt.hist(normal_data, bins=30)
plt.title("Normal Distribution")
plt.show()

plt.hist(right_skewed, bins=30)
plt.title("Right Skewed Distribution")
plt.show()

plt.hist(left_skewed, bins=30)
plt.title("Left Skewed Distribution")
plt.show()
print("Normal Data Mean:", np.mean(normal_data))
print("Normal Data Median:", np.median(normal_data))

print("Right Skewed Mean:", np.mean(right_skewed))
print("Right Skewed Median:", np.median(right_skewed))

print("Left Skewed Mean:", np.mean(left_skewed))
print("Left Skewed Median:", np.median(left_skewed))
```
