```python
#IQR method
import pandas as pd
data={
    "name":['A','B','C','D','E','F'],
    "age":[13,24,25,120,23,22]

}

df=pd.DataFrame(data)
q1=df['age'].quantile(0.25)
q3=df['age'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliners=df[(df['age']<lower_bound) | (df['age']>upper_bound)]
print("outliners using IQR method")
print(outliners)
```
